```python
csv_path='https://www.toronto.ca/ext/open_data/catalog/data_set_files/2016_neighbourhood_profiles.csv'
df = pd.read_csv(csv_path,encoding='latin1')
df.head()
```

```python
df.head(10)
```

```python
Neighbors = list(df.columns.values)
Neighbor = Neighbors[5:]
print(Neighbor)
```

```python
dfToronto = pd.DataFrame(index=Neighbor, columns=["Population","Average_Income","Population_Change"])
dfToronto.head()
```

```python
for index, row in dfToronto.iterrows():
    dfToronto.at[index, 'Population'] = df[index][2]
    dfToronto.at[index, 'Average_Income'] = df[index][2354]
    dfToronto.at[index, 'Population Change'] = df[index][4]

dfToronto.sort_values('Average_Income')
```
